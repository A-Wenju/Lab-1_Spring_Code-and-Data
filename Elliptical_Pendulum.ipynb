```python
#import the necessary packages
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
```

```python
#import the data
time, x_disp, y_disp = np.loadtxt("Elliptical Pendulum Data.txt", unpack = True, skiprows = 1)
print('Number of data points =', len(time)) #print the number of data points
```

```python
#Visualize the above data
plt.plot(x_disp, y_disp, label = 'Data Points')
plt.xlabel('x (distance)')
plt.ylabel('y (distance)')
plt.title('Position Graph for a Elliptical Pendulum')
plt.legend()
plt.show()
```

```python
#This program finds all the turning points 
Turning_points_x = []  #create empty lists to store the coordinates of turning points
Turning_points_y = []

for i in range(1, len(time)-1):  #create a for loop for all the data points
    forward_slope = (y_disp[i+1] - y_disp[i])/(x_disp[i+1] - x_disp[i])  #calculate the forward and backward slope from each point
    backward_slope = (y_disp[i] - y_disp[i-1])/(x_disp[i] - x_disp[i-1])

    if forward_slope * backward_slope < 0:  #if the slope changes for a point, store the coordinate values
        Turning_points_x.append(x_disp[i])
        Turning_points_y.append(y_disp[i])
```

```python
#Visualize the turning points for different number of ellipses
Number_of_Ellipses = 3  #Set the number of ellipses for which you want to visualize the turning points
counter = 4 * Number_of_Ellipses  #Each ellipse has four turning points
plt.scatter(Turning_points_x[:counter], Turning_points_y[:counter])  #scatter the turning points accordingly
plt.xlabel('x (distance)')  #label the axes and title
plt.ylabel('y (distance)')
plt.title("Position of the turning points")
plt.show()
```

```python
#This program plots the major axes
No_of_Major_axes = 13  #Set the number of major axes which you want to visualize
counter = 2*No_of_Major_axes #We need two points for each major axis

for i in range(0,counter,2):  #create a for loop to plot major axes
    i_1 = 2*i+1   #go to the alternating indices and find the coordinates for major axes
    i_2 = 2*i+3

    #Plot the elements in the above indices
    plt.plot([Turning_points_x[i_1], Turning_points_x[i_2]], [Turning_points_y[i_1], Turning_points_y[i_2]], marker = "o")

plt.xlabel('x (distance)')  #label the axes and title
plt.ylabel('y (distance)')
plt.title("Major Axes")
plt.xlim(-300, 350)  #set the limits in both the axes
plt.ylim(-150, 300)
plt.show()
```
